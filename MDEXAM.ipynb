```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#2. Read the CSV File

data = pd.read_csv('/content/heart_2020_cleaned.csv')

data.head(10)
```

# 1).Physical Activity and Heart Disease Across Age Categories?

```python
age_physical_activity = data.groupby(['AgeCategory', 'PhysicalActivity'])['HeartDisease'].mean().unstack()


age_physical_activity.plot(kind='bar', stacked=True, figsize=(10,6), color=['BLUE', 'RED'])
plt.title('Heart Disease and Physical Activity Across Age Categories')
plt.xlabel('Age Category')
plt.ylabel('Proportion with Heart Disease')
plt.xticks(rotation=45)
plt.legend(title='Physical Activity', labels=['Inactive', 'Active'])
plt.show()
```

# 2).Does smoking affect the likelihood of having heart disease?

```python
data = pd.read_csv('/content/heart_2020_cleaned.csv')

data['HeartDisease'] = (data['HeartDisease'] == 'Yes').astype(int)
smoking_heart_disease = data.groupby('Smoking')['HeartDisease'].mean()
smoking_heart_disease.plot(kind='bar', figsize=(8,6), color=['lightblue', 'lightcoral'])

plt.title('Heart Disease by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Proportion of Heart Disease')
plt.show()
```

OBSERVATION: The bar chart compares the proportion of individuals with heart disease in two groups smokers and nonsmokers
